```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome()
driver.get("https://wuzzuf.net/jobs/p/bu1ksr9tke5x-data-scientist-lumin-giza-egypt?o=2")

time.sleep(5)

soup = BeautifulSoup(driver.page_source, 'html.parser')

job_name = soup.find('h1', class_='css-f9uh36').text
company_name = soup.find('a', class_='css-p7pghv').text
company_location = soup.find(string="Sheikh Zayed, Giza")
published_date = soup.find('span', class_='css-182mrdn').text
skills = soup.find('div', class_='css-qe7mba').text

print(f"""
Job Name: {job_name}

Company Name: {company_name}

Company Location: {company_location}

Published Date: {published_date}

Skills: {skills}
""")
```
